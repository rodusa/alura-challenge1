### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url, decimal=",")
loja2 = pd.read_csv(url2, decimal=",")
loja3 = pd.read_csv(url3, decimal=",")
loja4 = pd.read_csv(url4, decimal=",")

```

```python
# Verificar se valores de frete das Lojas são coerentes. Para isso, precisamos garantir que "Frete" esteja como número (float)

def formatar_frete(df):
    df["Frete"] = pd.to_numeric(df["Frete"], errors="coerce")
    frete_formatado = df["Frete"].map("{:,.2f}".format).str.replace(",", "X").str.replace(".", ",").str.replace("X", ".")
    if "Frete Formatado" not in df.columns:
        pos = df.columns.get_loc("Frete") + 1
        df.insert(pos, "Frete Formatado", frete_formatado)
    else:
        df["Frete Formatado"] = frete_formatado

# Aplicar para todas as lojas
formatar_frete(loja1)
formatar_frete(loja2)
formatar_frete(loja3)
formatar_frete(loja4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Adicionar coluna identificadora da loja
loja1["Loja"] = "Loja 1"
loja2["Loja"] = "Loja 2"
loja3["Loja"] = "Loja 3"
loja4["Loja"] = "Loja 4"

# Combinar todos os dados em um único DataFrame
todas_lojas = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Calcular faturamento total por loja
faturamento_por_loja = todas_lojas.groupby("Loja")["Preço"].sum().sort_values(ascending=False)

# Gráfico de Barras
plt.figure(figsize=(10, 6))
plt.bar(faturamento_por_loja.index, faturamento_por_loja.values)
plt.title('Faturamento Total por Loja (Gráfico de Barras)')
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
import seaborn as sns
# Agrupar por loja e categoria e contar quantidade de vendas
vendas_por_categoria = todas_lojas.groupby(["Loja", "Categoria do Produto"])["Produto"].count().reset_index()
vendas_por_categoria = vendas_por_categoria.rename(columns={"Produto": "Quantidade de Vendas"})

# Paleta de cores personalizada
palette_custom = {
    "Loja 1": "blue",     # Azul
    "Loja 2": "gold",     # Amarelo
    "Loja 3": "#2CA02C",  # Verde
    "Loja 4": "red",      # Vermelho
}

# Criar gráfico de barras agrupadas
plt.figure(figsize=(12, 6))
sns.barplot(data=vendas_por_categoria,
            x="Categoria do Produto",
            y="Quantidade de Vendas",
            hue="Loja",
            palette=palette_custom)

plt.title("Quantidade de Produtos Vendidos por Categoria e por Loja")
plt.xlabel("Categoria do Produto")
plt.ylabel("Quantidade de Vendas")
plt.xticks(rotation=45)
plt.legend(title="Loja")
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média das avaliações por loja
media_avaliacoes = todas_lojas.groupby("Loja")["Avaliação da compra"].mean().reset_index()
media_avaliacoes = media_avaliacoes.rename(columns={"Avaliação da compra": "Média de Avaliação"})

# Gráfico de Pizza
plt.figure(figsize=(8, 8))
plt.pie(media_avaliacoes["Média de Avaliação"],
        labels=media_avaliacoes["Loja"],
        autopct='%1.1f%%',
        startangle=90)
plt.title('Média das Avaliações por Loja')
plt.axis('equal')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Analisar os 10 produtos mais vendidos por loja
# Função para gerar gráfico de colunas do Top 10 produtos mais vendidos por loja
def plot_top_10_produtos(loja_nome):
    top_produtos = todas_lojas[todas_lojas["Loja"] == loja_nome]["Produto"].value_counts().head(10).reset_index()
    top_produtos.columns = ["Produto", "Quantidade de Vendas"]

    plt.figure(figsize=(10, 6))
    plt.bar(top_produtos["Produto"], top_produtos["Quantidade de Vendas"])
    plt.title(f"Top 10 Produtos Mais Vendidos - {loja_nome}")
    plt.xlabel("Produto")
    plt.ylabel("Quantidade de Vendas")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
# Gerar gráficos para Loja 1
plot_top_10_produtos("Loja 1")
```

```python
# Gerar gráficos para Loja 2
plot_top_10_produtos("Loja 2")
```

```python
# Gerar gráficos para Loja 3
plot_top_10_produtos("Loja 3")
```

```python
# Gerar gráficos para Loja 4
plot_top_10_produtos("Loja 4")
```

# 5. Frete Médio por Loja


```python
# Calcular o frete médio por loja
frete_medio = todas_lojas.groupby("Loja")["Frete"].mean().reset_index()
frete_medio = frete_medio.rename(columns={"Frete": "Frete Médio"})

# Gráfico de Pizza
plt.figure(figsize=(8, 8))
plt.pie(frete_medio["Frete Médio"],
        labels=frete_medio["Loja"],
        autopct='%1.1f%%',
        startangle=90)
plt.title("Frete Médio por Loja")
plt.axis('equal')
plt.show()
```
